````python
Q1- Discuss string slicing and provide examples?
Ans1- String slicing is a powerful feature in Python that allows you to extract specific parts of a string.

Syntax:

string[start:stop:step]

- start: The starting index (inclusive).
- stop: The ending index (exclusive).
- step: The increment between indices (default is 1).

Examples:

1. Basic Slicing:


my_string = "Hello, World!"
print(my_string[0:5])  # Output: "Hello"


1. Omitting Start Index:


my_string = "Hello, World!"
print(my_string[:5])  # Output: "Hello"


1. Omitting Stop Index:


my_string = "Hello, World!"
print(my_string[7:])  # Output: "World!"


1. Negative Indices:


my_string = "Hello, World!"
print(my_string[-5:])  # Output: "World"


1. Reversing a String:


my_string = "Hello, World!"
print(my_string[::-1])  # Output: "!dlroW ,olleH"

Q2- Explain the key features of lists in python?
Ans2- Lists are a fundamental data structure in Python, offering flexibility and versatility.

Key Features:

1. Ordered Collection: Lists maintain the order of elements.

2. Mutable: Lists can be modified after creation.

3. Indexed: Elements are accessed via zero-based indexing.

4. Dynamic Size: Lists can grow or shrink dynamically.

5. Heterogeneous: Lists can store diverse data types.

List Operations:

1. Indexing: Accessing elements using square brackets [].


```
my_list = [1, 2, 3]
print(my_list[0])  # Output: 1


2. Slicing: Extracting subsets using start:stop:step.

   python
my_list = [1, 2, 3, 4, 5]
print(my_list[1:4])  # Output: [2, 3, 4]


1. Append: Adding elements using append().


```
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

Q3- Describe how to acess, modify, and delete elements in a list with examples?
Ans3- Here's how to access, modify, and delete elements in a list:

Accessing Elements:

1. Indexing: Use square brackets [] with the index.


my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1


1. Negative Indexing: Use -1 to access the last element.


my_list = [1, 2, 3, 4, 5]
print(my_list[-1])  # Output: 5


1. Slicing: Use start:stop:step to access a range.


my_list = [1, 2, 3, 4, 5]
print(my_list[1:4])  # Output: [2, 3, 4]


Modifying Elements:

1. Assignment: Assign a new value to an index.


my_list = [1, 2, 3, 4, 5]
my_list[0] = 10
print(my_list)  # Output: [10, 2, 3, 4, 5]


1. Slice Assignment: Assign new values to a range.


my_list = [1, 2, 3, 4, 5]
my_list[1:4] = [20, 30, 40]
print(my_list)  # Output: [1, 20, 30, 40, 5]

Q4- Compare and contrast tuples and lists with examples?
Ans4- Tuples and lists are fundamental data structures in Python.

Similarities:

1. Ordered collections
2. Indexed (zero-based)
3. Can store multiple data types
4. Support slicing and indexing

Differences:

1. Immutability: Tuples are immutable, while lists are mutable.
2. Syntax: Tuples use parentheses (), while lists use square brackets [].
3. Performance: Tuples are faster and more memory-efficient.
4. Methods: Lists have more methods (e.g., append, insert, remove).

Tuple Characteristics:

1. Immutable
2. Faster execution
3. Less memory usage
4. Useful for constant data

List Characteristics:

1. Mutable
2. Slower execution
3. More memory usage
4. Useful for dynamic data

Examples:

Tuple:


my_tuple = (1, 2, 3)
print(my_tuple[0])  # Output: 1
# my_tuple[0] = 10  # Error: Tuples are immutaâ€¦

Q5- Describe the key features of sets and provide examples of their use?
Ans5- Sets are an essential data structure in Python, providing efficient and convenient ways to store and manipulate unique elements.

Key Features:

1. Unordered collection
2. Unique elements (no duplicates)
3. Mutable (can be modified)
4. Fast membership testing
5. Support for mathematical operations (union, intersection, difference)

Creating Sets:

1. Using the set() function:

my_set = set([1, 2, 3, 4, 5])

1. Using curly brackets {}:

my_set = {1, 2, 3, 4, 5}

1. From a list or tuple:

my_list = [1, 2, 2, 3, 4, 4, 5]
my_set = set(my_list)


Set Operations:

1. Union (|): Combines two sets.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}

Q6- Discuss the use cases of tuples and sets in python programming?
Ans6- Tuples and sets are essential data structures in Python, serving various purposes.

Tuple Use Cases:

1. Constant data: Store data that shouldn't change.
2. Function arguments: Pass multiple arguments efficiently.
3. Dictionary keys: Use tuples as keys due to their immutability.
4. Data integrity: Ensure data consistency and accuracy.
5. Performance-critical code: Tuples are faster than lists.
6. Database records: Represent rows or records.
7. Geometric coordinates: Store x, y, z coordinates.
8. Color representations: Store RGB or RGBA values.

Set Use Cases:

1. Duplicate removal: Eliminate duplicates from lists.
2. Fast membership testing: Check element presence efficiently.
3. Data deduplication: Remove duplicate data.
4. Intersection, union, difference: Perform set operations.
5. Data validation: Check data against a set of valid values.
6. Unique identifier generation: Use sets to generate unique IDs.
7. Graph algorithms: Represent nodes and edges.
8. Data analysis: Use sets for filtering, grouping, and aggregating.

Real-World Examples:

Tuples:

1. GPS coordinates (latitude, longitude)
2. Date and time representations (year, month, day, hour, minute, second)
3. RGB color values (red, green, blue)
4. Database records (id, name, email, phone)

Sets:

1. Unique user IDs
2. Product categories (e.g., electronics, clothing)
3. Valid input values (e.g., countries, languages)
4. Friends or followers in social media platforms

Benefits:

Tuples:

1. Immutability ensures data integrity
2. Faster execution compared to lists
3. Memory-efficient

Sets:

1. Fast membership testing
2. Efficient duplicate removal
3. Simplified data analysis

Best Practices:

Tuples:

1. Use meaningful variable names
2. Avoid unnecessary conversions
3. Use tuples for constant data

Sets:

1. Use meaningful variable names
2. Choose sets for fast membership testing
3. Avoid using sets for ordered data

By understanding the use cases and benefits of tuples and sets, you'll write more efficient, effective, and Pythonic code.

Q7- Describe how to add,modify, and delete items in a dictionary with examples?
Ans7- Dictionaries in Python are mutable data structures that store mappings of unique keys to values.

Adding Items:

1. Direct Assignment:


my_dict = {"name": "John", "age": 30}
my_dict["city"] = "New York"
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York"}


1. update() Method:


my_dict = {"name": "John", "age": 30}
my_dict.update({"city": "New York", "country": "USA"})
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York", "country": "USA"}


1. setdefault() Method:


my_dict = {"name": "John", "age": 30}
my_dict.setdefault("city", "New York")
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York"}
Q8- Discuss the importance of dictionary keys being immutable and provide examples?
Ans8- In Python, dictionary keys must be immutable, meaning they cannot be changed after creation.

Why Immutability Matters:

1. Hashing: Dictionaries use hashing to store and retrieve key-value pairs efficiently. Immutable keys ensure consistent hash values.
2. Uniqueness: Immutable keys prevent accidental key modifications, ensuring unique keys.
3. Performance: Immutable keys optimize dictionary lookups and insertions.

Immutable Key Types:

1. Strings (e.g., "hello")
2. Integers (e.g., 123)
3. Floats (e.g., 3.14)
4. Tuples (e.g., (1, 2, 3))
5. Boolean values (e.g., True, False)
6. None

Mutable Key Types (Not Allowed):

1. Lists (e.g., [1, 2, 3])
2. Dictionaries (e.g., {"a": 1})
3. Sets (e.g., {1, 2, 3})

Examples:

Valid Dictionary with Immutable Keys:

my_dict = {"name": "John", 123: "age", ("a", "b"): "tuple_key"}
print(my_dict)  # Output: {'name': 'John', 123: 'age', ('a', 'b'): 'tuple_key'}


Invalid Dictionary with Mutable Keys:

my_list = [1, 2, 3]
my_dict = {my_list: "value"}  # TypeError: unhashable type: 'list'


Converting Mutable Keys to Immutable:

my_list = [1, 2, 3]
my_tuple = tuple(my_list)  # Convert list to tuple
my_dict = {my_tuple: "value"}  # Valid dictionary
````
