Q1- What are the five key concepts of object- oriented programming (OOP)?
Ans1- The five key concepts of Object-Oriented Programming (OOP) are:

1. Encapsulation

Encapsulation is the concept of bundling data and methods that operate on that data within a single unit, called a class or object. This helps to hide the implementation details of an object from the outside world and provides a way to access and modify the data only through controlled interfaces.

2. Abstraction

Abstraction is the concept of representing complex systems in a simplified way by exposing only the necessary information to the outside world. Abstraction helps to reduce complexity and improve modularity by hiding the implementation details of an object and showing only its essential features.

3. Inheritance

Inheritance is the concept of creating a new class based on an existing class. The new class, called the subclass or derived class, inherits the properties and behavior of the existing class, called the superclass or base class. Inheritance helps to promote code reuse and facilitates the creation of a hierarchy of related classes.

4. Polymorphism

Polymorphism is the concept of taking on multiple forms. In OOP, polymorphism refers to the ability of an object to take on multiple forms, depending on the context in which it is used. Polymorphism can be achieved through method overloading, method overriding, or operator overloading.

5. Composition

Composition is the concept of combining objects or data types to form a new object or data type. Composition helps to create complex objects from simpler ones and facilitates the creation of objects that have a hierarchical or aggregate structure.

These five key concepts of OOP provide the foundation for designing and developing robust, maintainable, and scalable software systems.

Q3- Explain the difference between instance methods and class methods. Provide an example of each ?
Ans3- In Python, instance methods and class methods are two types of methods that can be defined in a class. The main difference between them is the way they are called and the data they have access to.

Instance Methods

Instance methods are methods that belong to an instance of a class. They are called on an instance of the class and have access to the instance's attributes. Instance methods are used to perform actions that are specific to an instance of the class.

Example:

class Car:
    def __init__(self, color):
        self.color = color

    def honk(self):
        print(f"The {self.color} car honks!")

my_car = Car("red")
my_car.honk()  # Output: The red car honks!

In this example, the honk method is an instance method because it is called on an instance of the Car class (my_car). The method has access to the instance's color attribute.

Class Methods

Class methods are methods that belong to a class itself, rather than an instance of the class. They are called on the class and have access to the class's attributes. Class methods are used to perform actions that are related to the class as a whole, rather than a specific instance.

Example:

class Car:
    num_cars = 0

    def __init__(self, color):
        self.color = color
        Car.num_cars += 1

    @classmethod
    def get_num_cars(cls):
        return cls.num_cars

my_car = Car("red")
print(Car.get_num_cars())  # Output: 1

In this example, the get_num_cars method is a class method because it is called on the Car class itself, rather than an instance of the class. The method has access to the class's num_cars attribute.

Note the use of the @classmethod decorator to indicate that the get_num_cars method is a class method. The cls parameter is used to refer to the class itself, rather than an instance of the class.

Q6- What is the method resolution order ( MRO) in python ? How can you retrieve it programmatically?
Ans6-[9:45 PM, 12/5/2024] Meta AI: In Python, the Method Resolution Order (MRO) is the order in which Python searches for methods and attributes in a class's inheritance hierarchy. It is used to resolve the order of method calls when there are multiple classes in the inheritance chain.

The MRO is based on the C3 linearization algorithm, which ensures that the order of method calls is consistent and predictable.

Here are the key points about the MRO in Python:

1. Depth-first search: Python searches for methods and attributes in the current class first, then in its parent classes, and so on.
2. Left-to-right search: When there are multiple parent classes, Python searches them from left to right.
3. Avoiding ambiguity: The MRO ensures that there is no ambiguity when resolving method calls.

To retrieve the MRO programmatically in Python, you can use the mro() method of a class. Here's an example:


class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

print(D.mro())


Output:


(<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)


In this example, the mro() method returns a tuple containing the classes in the MRO of the D class. The order of the classes in the tuple reflects the order in which Python searches for methods and attributes in the D class's inheritance hierarchy.

def timer_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time

        return result
    return wrapper

# Example usage

Q11- Explain the concept of the diamond problem in multiple inheritance. How does python resolve it ?
ANS11- The diamond problem is a classic issue that arises in multiple inheritance, where a class inherits from two or more classes that have a common base class. This creates a diamond-shaped inheritance graph, where the base class is at the top, and the two intermediate classes are at the middle, with the derived class at the bottom.

Here's an example of the diamond problem in Python:


class A:
    def method(self):
        print("A's method")

class B(A):
    def method(self):
        print("B's method")

class C(A):
    def method(self):
        print("C's method")

class D(B, C):
    pass

d = D()
d.method()  # Which method will be called?


In this example, class D inherits from both B and C, which both inherit from A. This creates a diamond-shaped inheritance graph. The question is, which method() will be called when we invoke d.method()?

Python resolves the diamond problem using a technique called C3 linearization, also known as the "Method Resolution Order" (MRO). The MRO is a list of classes that Python searches in order to find a method or attribute.

In Python 3.x, the MRO is calculated using the following rules:

1. List the classes in the order they are inherited from, from left to right.
2. If a class is already in the list, skip it.
3. If a class is not in the list, add it and all its parents (except for object) to the list.

Using these rules, we can calculate the MRO for class D:

1. Start with D.
2. Add B and its parents (A and object).
3. Add C and its parents (A and object), but skip A since it's already in the list.

The resulting MRO for class D is: [D, B, C, A, object].

When we invoke d.method(), Python searches the MRO for a class that defines a method() attribute. It finds B first, so it calls B's method().

Here's the corrected code with the MRO printed out:

    def method(self):
        print("A's method")

        print("B's method")

        print("C's method")

    pass
